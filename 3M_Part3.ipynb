#### *Break-Pass and Continue*

```python
total = 5
x = int(input())
i=1
while(i<=x):
    print("Chocolate")
    i+=1
```

```python
#Break statement
avail = 10
x = int(input())
i=1
while(i<=x):
    if i>avail:
        break
    print("Chocolate")
    i+=1
print("Exausted")
```

### Functions

```python
def add():
    a = int(input())
    b = int(input())
    c = a+b
    return c
add()
```

```python
def add(a,b):
    c=a+b
    return c
add(10,10)
```

```python
'''Print the below output using functions
0 0
1 1
2 4
3 9
4 16 
'''

def square(val):
    for i in range(val):
        print(i,i*i)
square(5)
```
