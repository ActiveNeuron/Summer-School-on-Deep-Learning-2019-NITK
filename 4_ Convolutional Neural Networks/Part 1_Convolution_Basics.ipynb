### Convolution and Correlation

```
from scipy.ndimage import correlate,convolve
import numpy as np


a = np.array([1,2,3,4,5])

b = np.array([1,2,3])

c = convolve(a,b)
print(c)

d = correlate(a,b)
print(d)
```

```
from scipy.signal import convolve2d
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np

img = cv2.imread('cam.png',cv2.IMREAD_GRAYSCALE)


kernel = np.array([[1,0,-1],
                   [0,0,0],
                   [-1,0,1]])


print(img.shape)
print(kernel.shape)

b=convolve2d(img,kernel)
```

```
plt.figure(figsize=(10,10))
plt.subplot(131).imshow(img,'gray')
plt.subplot(132).imshow(kernel,'gray')
plt.subplot(133).imshow(b,'gray')
```

```
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
from skimage import filters


img = cv2.imread('cam.png',0)
plt.imshow(img)
print(img.shape)
```

```
h = filters.sobel_h(img)
v = filters.sobel_v(img)
res1 = filters.sobel(img)
```

```
plt.figure(figsize=(20,20))
plt.subplot(141).imshow(img,'gray')
plt.subplot(142).imshow(h,'gray')
plt.subplot(143).imshow(v,'gray')
plt.subplot(144).imshow(res1,'gray')
```
